## Question 1a

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math

p1 = 0.5
p2 = 0.5
mu1 = -5
mu2 = -10
variance = 5
sigma = math.sqrt(variance)

size = 10000

# w1 = np.linspace(mu1 - 3*sigma, mu1 + 3*sigma, size)
# w2 = np.linspace(mu2 - 3*sigma, mu2 + 3*sigma, size)

w = np.linspace(-20,5,size)

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4])

px_1 = mlab.normpdf(w, mu1, sigma);
px_2 = mlab.normpdf(w, mu2, sigma);

ax1.plot(w,px_1)
ax1.plot(w,px_2)

plt.show()
```

\begin{align}
g_i(x) & = -\frac{1}{2}\log{2\pi} - \log{\sigma_i} - \frac{(x-m_i)^2}{2\sigma^2_i} + \log{P(C_i)}\\
\end{align}

To find separation surface: $$ $$
\begin{align}
g_1(x) &= g_2(x)\\
-\frac{1}{2}\log{2\pi} - \log{\sigma} - \frac{(x-m_1)^2}{2\sigma^2} + \log{P(C_1)} & = -\frac{1}{2}\log{2\pi} - \log{\sigma} - \frac{(x-m_2)^2}{2\sigma^2} + \log{P(C_2)}\\
-(x-m_1)^2 & = -(x-m_2)^2\\
x & = \frac{m_1 + m_2}{2}\\
x & = \frac{-5 + -10}{2}\\
x & = -7.5
\end{align}

```python
fig = plt.figure()
ax2 = fig.add_axes([0.1, 0, 0.8, 0.4])

px = px_1*p1+px_2*p2;
p1_r = px_1*p1/px;
p2_r = px_2*p2/px;

ax2.plot(w,p1_r);
ax2.plot(w,p2_r);

sep = np.linspace(0,1,size);
ax2.plot(np.full(size,-7.5),sep)

plt.show()
```

## Question 1c (a)

```python
p1 = 0.5
p2 = 0.5
binSize = 100;

fig = plt.figure()
ax5 = fig.add_axes([0.1, 0.5, 0.8, 0.4])

# data set (s1 + s2) is created with equal probability 
s1 = np.random.normal(mu1, sigma, (int)(size*p1))
s2 = np.random.normal(mu2, sigma, (int)(size*p2))

count1, bins1, ignored1 = plt.hist(s1, binSize, normed=True)
count2, bins2, ignored2 = plt.hist(s2, binSize, normed=True)

bins = np.concatenate((bins1, bins2), axis=0);
bins.sort()

#p1_dist = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins1 - mu1)**2 / (2 * sigma**2) );
#p2_dist = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins2 - mu2)**2 / (2 * sigma**2) );
p1_dist = mlab.normpdf(bins, mu1, sigma);
p2_dist = mlab.normpdf(bins, mu2, sigma);

ax5.plot(bins, p1_dist,linewidth=2, color='b')
ax5.plot(bins, p2_dist,linewidth=2, color='r')

plt.show()
```

```python
r1_mean = print('P1 mean %d' % np.mean(s1))
r2_mean = print('P2 mean %d' % np.mean(s2))
r1_var = print('P1 var %d' % np.var(s1))
r2_var = print('P2 var %d' % np.var(s2))
```

To find separation surface: $$ $$
\begin{align}
g_1(x) &= g_2(x)\\
-\log{\sigma_1} - \frac{(x-m_1)^2}{2\sigma_1^2}  & = -\log{\sigma_2} - \frac{(x-m_2)^2}{2\sigma_2^2} \\
(x-m_1)^2  & = (x-m_2)^2 \\
(x+5)^2  & = (x+10)^2 \\

\end{align}

```python
# from IPython.core.debugger import Tracer; Tracer()() 
fig = plt.figure()
ax6 = fig.add_axes([0.1, 0.5, 0.8, 0.4])

px_3 = p1_dist*p1+p2_dist*p2;
p1_r = p1_dist*p1/px_3;
p2_r = p2_dist*p2/px_3;
#w1 = np.linspace(-20,5,size)

ax6.plot(bins,p1_r);
ax6.plot(bins,p2_r);

ax6.plot(np.full(size,-7.5),np.linspace(0,1,size));

plt.show()
```
